```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  
        self.__balance = balance  

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance
    
account = BankAccount("Bohdan", 1000)
print(f"баланс на старті: {account.get_balance()}\n")

for i in range(1, 6):
    rand_deposit = random.randint(50, 600)
    rand_withdraw = random.randint(50, 600)
    
    account.deposit(rand_deposit)
    print(f"Ітерація {i}: Поповнено на {rand_deposit}. Поточний баланс: {account.get_balance()}")
    
    result = account.withdraw(rand_withdraw)
    if result == "Insufficient funds":
        print(f"Ітерація {i}: Спроба зняти {rand_withdraw}. Відмова (недостатньо коштів).")
    else:
        print(f"Ітерація {i}: Успішно знято {result}. Поточний баланс: {account.get_balance()}")
    
    print("-" * 30)

print(f"Кінцевий результат: рахунок користувача {account.owner} становить {account.get_balance()}")
```
