```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  
        self.__balance = balance  

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            return amount
        else:
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance
    
account = BankAccount("Bohdan", 1000)
print(f"баланс на старті: {account.get_balance()}\n")

for i in range(1, 6):
    rand_deposit = random.randint(50, 600)
    rand_withdraw = random.randint(50, 600)
    
    account.deposit(rand_deposit)
    print(f"Ітерація {i}: Поповнено на {rand_deposit}. Поточний баланс: {account.get_balance()}")
    
    result = account.withdraw(rand_withdraw)
    if result == "Insufficient funds":
        print(f"Ітерація {i}: Спроба зняти {rand_withdraw}. Відмова (недостатньо коштів).")
    else:
        print(f"Ітерація {i}: Успішно знято {result}. Поточний баланс: {account.get_balance()}")
    
    print("-" * 30)

print(f"Кінцевий результат: рахунок користувача {account.owner} становить {account.get_balance()}")
```

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"
    
    def start_engine(self):
        return f"Двигун транспорту {self.brand} запущений"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model) 
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

car = Car("Toyota", "Camry", 5)
print(car.display_info())

print(car.start_engine())
```

```python
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Гав"

class Cat(Animal):
    def speak(self):
        return "Мяу:3"

class Fish(Animal):
    pass

animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(f"{type(animal).__name__} скаже зараз: {animal.speak()}")
```

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=100):
        self.name = name
        self.health = health

    @abstractmethod
    def attack(self, another_item):
        pass

    @abstractmethod
    def buff(self):
        pass


class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power 
        self._sharp = 0 

    def attack(self, another_item: Item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"Удар мечиком, нанесено {self.name} ({current_attack} урона). У {another_item.name} залишилось {another_item.health} хепешки."

    def buff(self):
        self._sharp += 2
        return f"Меч {self.name} прокачено. Додаткова гострота: {self._sharp}"


class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._weight_bonus = 0

    def attack(self, another_item: Item):
        current_attack = self.__attack_power + self._weight_bonus + randint(0, 20)
        another_item.health -= current_attack
        return f"Удар сокиркою, нанесено {self.name} ({current_attack} урона). У {another_item.name} залишилось {another_item.health} хепешки."

    def buff(self):
        self._weight_bonus += 3
        return f"Сокиру {self.name} розігнано, Бонус: {self._weight_bonus}"


class Bow(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._range_power = 0 

    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(5, 15) + self._range_power
        another_item.health -= current_attack
        return f"Постріл з лука, нанесено {self.name} ({current_attack} шкоди). У {another_item.name} залишилось {another_item.health} хепешки."

    def buff(self):
        self._range_power += 1
        return f"Лук {self.name} перезаряджено(типу), Дальність/точність збільшено: {self._range_power}"


def start_game():
    print("Вітаємо на кровавій арені, ми давно не влаштовували жорстоких боїв тут")
    
    weapon_classes = [
        lambda name: Sword(name, 10),
        lambda name: Axe(name, 12),
        lambda name: Bow(name, 8)
    ]
    
    player = choice(weapon_classes)("Гравець")
    enemy = choice(weapon_classes)("Суперник")

    print(f"Моя зброя: {type(player).__name__}")
    print(f"Зброя супротивника: {type(enemy).__name__}")
    print("=" * 40)

    round_num = 1

    while player.health > 0 and enemy.health > 0:
        print(f"\n--- РАУНД {round_num} ---")
        print(f"Моя хепешка: {player.health} |  хепешка суперника: {enemy.health}")
        
        print("1. Атакувати")
        print("2. Підсилити зброю")
        action = input("Оберіть дію (1 або 2): ")
        
        print("\n[Ваш хід]")
        if action == '1':
            print(player.attack(enemy))
        elif action == '2':
            print(player.buff())
        else:
            print("Пропуск ходу просто і все")

        if enemy.health <= 0:
            print("\nПЕРЕМОГА БУДЕ")
            break

        print("\n[Хід супротивника]")
        enemy_action = randint(1, 2)
        if enemy_action == 1:
            print(enemy.attack(player))
        else:
            print(enemy.buff())

        if player.health <= 0:
            print("\nТи слився бот")
            break
        
        round_num += 1

start_game()
```
